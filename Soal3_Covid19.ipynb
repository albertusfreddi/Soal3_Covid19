```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
import folium as fo
```

```python
# Scraping angka

url = 'https://en.wikipedia.org/wiki/2020_coronavirus_pandemic_in_Indonesia'
html = requests.get(url)
soup = BeautifulSoup(html.content, 'html.parser')

x = soup.find_all('tbody')
listSementara = x[3].text

listSementara = listSementara.split('\n')
listDataText = []
listDataAngka = []
for i in listSementara:
    if len(i)>0:
        if i[0] in ['1','2', '3', '4', '5', '6', '7', '8', '9', '0']:
            listDataAngka.append(i)
        else:
            listDataText.append(i)
            

listDataAngkaConfirmed = listDataAngka[4:140:4]
listDataAngkaRecovered = listDataAngka[5:140:4]
listDataAngkaDeaths = listDataAngka[6:140:4]
listDataAngkaActive = listDataAngka[7:140:4]

listDataWeb = []
listDataTextBaru = []
for i in listDataText:
    if '.' in i:
        listDataWeb.append(i)
    else :
        listDataTextBaru.append(i.replace('[b]',''))
listDataProv = listDataTextBaru[6:40]
```

```python
# Scrap data latitute, longitude dan logo

url = 'https://raw.githubusercontent.com/LintangWisesa/Indonesia-Covid19-Maps/master/data/gps_indonesia.json'
data = requests.get(url).json()

listLatitude = []
listLongitude = []
listLogo = []

for i in data:
    listLatitude.append(i['latitude'])
    listLongitude.append(i['longitude'])
    listLogo.append(i['logo'])
```

```python
# Membuat dataframe dari data yang telah di cleaning

df = pd.DataFrame(list(zip(listDataProv, listDataAngkaConfirmed, listDataAngkaRecovered, listDataAngkaDeaths, listDataAngkaActive,
                          listLatitude, listLongitude, listLogo)), 
               columns =['Provinsi', 'Confirmed', 'Recovered', 'Deaths', 'Active', 'Latitude', 'Longitude', 'Logo'])
df
```

```python
peta = fo.Map(
    location = [-3.1252829,116.8325654],
    zoom_start = 4.7
)

for i in range(len(df)):
    popups = fo.Popup('Confirmed : {} \nRecovered : {} \nDeaths : {}'.format(df['Confirmed'][i], df['Recovered'][i], df['Deaths'][i]), max_width=300,min_width=100)
    icon = fo.features.CustomIcon(df['Logo'][i],icon_size=(28, 30))
    fo.Marker([df['Latitude'][i],df['Longitude'][i]],
          popup = popups,
        tooltip = '<i>{}</i>'.format(df['Provinsi'][i]),
         icon = icon).add_to(peta)


peta.save('petaCovid-19.html')
peta
```
